## Importar as bibliotecas

```python
# Importar as bibliotecas
import numpy as np
import pandas as pd
```

## Criando o "dataset do caos"

```python
# Criar o conjunto de dados com dados faltantes
dados_sujos = {
    'ID_Pedido': [101, 102, 103, 103, 105, 106, 107, 108, 109, 110],
    'Cliente': ['Ana', 'Bruno', 'Carlos', 'Carlos', 'Eva',
                'Frank', np.nan, 'Hannah', 'Ítalo', 'Julia'],
    'Valor': [150.50, np.nan, 200.00, 200.00, 50.00,
              -999, 120.00, np.nan, 300.00, 450.00],
    'Status': ['Pago', 'Pendente', 'Pago', 'Pago', 'pago',
               'Cancelado', 'Pendente', 'Pendente', 'Enviado', 'Enviado'],
    'Data': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-03', '2023-01-05',
             np.nan, '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10']
}

df = pd.DataFrame(dados_sujos)
print('Dataset "sujo" carregado')
df
```

## Detectando o inimigo: valores nulos (`NaN`)

```python
# Contagem de nulos por coluna
print('Valores nulos por coluna:')
df.isnull().sum()
```

```python
# Porcentagem de dados faltantes
percentual_nulos = (df.isnull().sum() / len(df)) * 100
print(f'Percentual de nulos:\n{percentual_nulos}')
```

## Tratamento de nulos: `dropna` e `fillna`

### `dropna()`: a opção nuclear

```python
# Remover as linhas que possuem valores NaN
df_limpo = df.dropna(
    axis=0,
    how='any',  # 'any': ao menos um NaN na linha / 'all': todas as células com NaN
)
df_limpo
```

```python
# Remover apenas se o nulo estiver em uma coluna específica
df_cliente_ok = df.dropna(subset=['Cliente'])
df_cliente_ok
```

### `fillna()`: a opção cirúrgica

```python
# Preencher o valor nulo com a média (estratégia comum)
media_valor = round(df['Valor'].mean(), 2)
df['Valor'] = df['Valor'].fillna(media_valor)
df
```

```python
# Preencher com uma constante
df['Cliente'] = df['Cliente'].fillna('Desconhecido')
df
```

```python
# Preencher com o valor da linha anterior (ótimo para séries temporais)
df['Data'] = df['Data'].fillna(method='ffill')
df
```

## Removendo duplicatas

```python
# Vereficaar se há duplicatas
print(f'Total de duplicatas: {df.duplicated().sum()}')
```

```python
# Remover duplicatas mantendo apenas a primeira ocorrência
df = df.drop_duplicates(keep='first')
df
```

## Substituição e padronização: `replace`

```python
# Corrigir valores impossíveis
df['Valor'] = df['Valor'].replace(-999, 0)
df
```

```python
# Padronizar os valores de uma coluna
df['Status'] = df['Status'].replace('pago', 'Pago')
df
```

```python
# Substituição múltipla usando dicionário
correcoes = {
    'Pendente': 'Aguardando',
    'Enviado': 'Em trânsito'
}
df['Status'] = df['Status'].replace(correcoes)
df
```

|Ferramenta|Quando usar|
|:-:|:-:|
|`isnull().sum()`|Para diagnosticar onde estão os valores `NaNs`|
|`dropna()`|Quando você tem muitos dados e pode se dar ao luxo de perder linhas|
|`fillna()`|Quando perder dados causa viés; usa-se média, mediana ou moda|
|`drop_duplicates()`|Essencial após unir tabelas ou ler logs de sistemas|
|`replace()`|Para limpar erros de digitação ou valores sentinela|
