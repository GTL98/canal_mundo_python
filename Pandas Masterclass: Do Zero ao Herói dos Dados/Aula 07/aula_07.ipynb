## Criando o Dataset

```python
# --- Importar o Pandas --- #
import pandas as pd

# --- Criar os dados --- #
dados = {
    'regiao': ['Norte', 'Norte', 'Sul', 'Sul', 'Norte', 'Sul', 'Oeste', 'Oeste'],
    'vendedor': ['Ana', 'Bruno', 'Carlos', 'Diego', 'Ana', 'Diego', 'Erika', 'Erika'],
    'produto': ['Celular', 'Tablet', 'Celular', 'Notebook', 'Notebook', 'Tablet', 'Celular', 'Monitor'],
    'vendas': [2500, 1200, 2500, 4500, 3800, 1300, 2600, 1500],
    'comissao': [125, 60, 125, 225, 190, 65, 130, 75]
}

# --- Criar o DataFrame --- #
df = pd.DataFrame(dados)
df
```

## Groupby Básico: Agregação Simples

```python
# --- Qual o valor total de vendas por região? --- #
vendas_regiao = df.groupby('regiao')['vendas'].sum()
vendas_regiao
```

```python
# --- Transformar o agrupamento em um DataFrame --- #
vendas_regiao_df = df.groupby('regiao')[['vendas']].sum()
vendas_regiao_df
```

## O Poder do `.agg()`: Agregações Customizadas

```python
# --- Aplicar múltiplas funções a uma coluna --- #
resumo_vendas = df.groupby('regiao')['vendas'].agg(['sum', 'mean', 'count', 'max'])
resumo_vendas
```

### Agregações Diferentes por Coluna

```python
# --- Separar as agregações por coluna --- #
dic_agg = {
    'vendas': 'sum',
    'comissao': 'mean'
}

resumo_especifico = df.groupby('regiao').agg(dic_agg)
resumo_especifico
```

## Agregação Nomeada

```python
# --- Nomear as colunas agrupadas --- #
resumo_vendedor = df.groupby('vendedor').agg(
    total_vendido=('vendas', 'sum'),
    media_comissao=('comissao', 'mean'),
    qtd_produtos=('produto', 'count')
)

resumo_vendedor.sort_values(by='total_vendido', ascending=False)
```

## Agrupamento por Múltiplas Colunas

```python
# --- Agrupar os dados por mais colunas --- #
multi_grupo = df.groupby(['regiao', 'produto'])[['vendas']].sum()
multi_grupo
```

|Técnica|Objetvo|
|:-:|:-:|
|`df.groupby('A')['B'].sum()`|Agregação mais simples e rápida|
|`.agg(['mean', 'std'])`|Várias estatísticas para a mesma coluna|
|`.agg({'A': 'sum', 'B': 'max'})`|Funções diferentes para colunas diferentes|
|**Agregação Nomeada**|O jeito mais limpo de gerar relatórios com nomes de colunas customizados|
