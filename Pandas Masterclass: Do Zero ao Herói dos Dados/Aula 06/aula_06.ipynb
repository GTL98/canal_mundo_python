## Criar o dataset

```python
# --- Importar o Pandas --- #
import pandas as pd

# --- Criar os dados --- #
dados = {
    'Cod_Produto': ['TECH-001', 'SOFT-002', 'TECH-003', 'GEAR-004'],
    'Descricao': [' notebook gamer v1', 'LICENÇA ANTIVÍRUS',
                  'monitor 4k ultra', ' MOUSE SEM FIO '],
    'Preco_Unitario': ['1500.00', '299.99', '3200.00', '150.00'],
    'Quantidade': [10, 50, 5, 100],
    'Custo_Percentual': [0.7, 0.2, 0.6, 0.4]
}

df = pd.DataFrame(dados)
df
```

```python
# --- Ver as informações dos dados --- #
df.info()
```

## Conversão de tipos: o poder do `astype`

```python
# --- Converter de string para float --- #
df['Preco_Unitario'] = df['Preco_Unitario'].astype(float)

# --- Verificar a mudança --- #
print(f'Novo tipo da coluna "Preco_Unitario": {df["Preco_Unitario"].dtype}')
```

## Operações aritméticas (vetorização)

```python
# --- Calcular a receita total --- #
df['Receita_Total'] = df['Preco_Unitario'] * df['Quantidade']

# --- Calcular o custo real --- #
df['Custo_Total'] = df['Receita_Total'] * df['Custo_Percentual']

# --- Calcular a margem de lucro --- #
df['Lucro_Liquido'] = df['Receita_Total'] - df['Custo_Total']

df[['Cod_Produto', 'Receita_Total', 'Lucro_Liquido']]
```

## Manipulação de strings com `.str`

```python
# --- Remover os espaços em branco e padronizar a escrita --- #
df['Descricao'] = df['Descricao'].str.strip().str.title()

# --- Criar uma coluna de Categoria a partir de Cod_Produto --- #
# --- O expland=True transforma o resultado em colunas separadas --- #
df['Categoria'] = df['Cod_Produto'].str.split('-', expand=True)[0]

# --- Verificara se há uma substring --- #
df['Eh_Gamer'] = df['Descricao'].str.contains('gamer', case=False)
df[['Cod_Produto', 'Categoria', 'Descricao', 'Eh_Gamer']]
```

## Métodos de string essenciais

|Método|O que faz
|:-:|:-:|
|`str.lower()`/`.upper()`|Muda a caixa do texto.|
|`str.replace('A', 'B')`|Substitui caracteres|
|`str.len()`|Conta o número de caracteres|
|`.str.slice(0, 3)`|Corta a string|
