## Criando os dados

```python
# --- Importar o Pandas --- #
import pandas as pd

# --- Tabela A: Clientes --- #
df_clientes = pd.DataFrame({
    'ID_Cliente': [1, 2, 3, 4],
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Diego'],
    'Estado': ['SP', 'RJ', 'MG', 'BA']
})

# --- Tabela B: Pedidos --- #
df_pedidos = pd.DataFrame({
    'Cod_Cliente': [1, 2, 2, 3, 5],
    'Item': ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Cadeira Gamer'],
    'Valor': [5000, 150, 300, 1500, 1200]
})
```

```python
df_clientes
```

```python
df_pedidos
```

## O método `merge()` e os tipos de cruzamento

### Inner Join (interceção)
Traz apenas o que existe em **ambas** as tabelas

```python
# --- left_on e right_on são usados porque as colunas têm nomes diferentes --- #
df_inner = pd.merge(
    left=df_clientes,
    right=df_pedidos,
    left_on='ID_Cliente',
    right_on='Cod_Cliente',
    how='inner'
)
df_inner
```

## Left Join (foco na esquerda)
Traz **todos** os clientes da tabela da esquerda (Clientes). Se alguém não comprou nada (caso do diego), os campos de pedido viram `NaN`

```python
df_left = pd.merge(
    left=df_clientes,
    right=df_pedidos,
    left_on='ID_Cliente',
    right_on='Cod_Cliente',
    how='left'
)
df_left
```

### Right Join (foco na direita)
Traz **todos** os pedidos da tabela da direita. O pedido do cliente ID 5 aparece, mas o nome e o estado dele ficam vazios porque ele não está no cadastro

```python
df_right = pd.merge(
    left=df_clientes,
    right=df_pedidos,
    left_on='ID_Cliente',
    right_on='Cod_Cliente',
    how='right'
)
df_right
```

### Outer Join (união total)
Traz absolutamente **tudo** de ambas as tabelas, preenchendo com `NaN` onde não houver correspondência

```python
df_outer = pd.merge(
    left=df_clientes,
    right=df_pedidos,
    left_on='ID_Cliente',
    right_on='Cod_Cliente',
    how='outer'
)
df_outer
```

## O método `concat()`: empilhamento

```python
# --- Simular dados de venda de meses diferentes --- #
vendas_jan = pd.DataFrame({
    'Produto': ['A', 'B'],
    'Qtd': [10, 20]
})
vendas_fev = pd.DataFrame({
    'Produto': ['C', 'A'],
    'Qtd': [15, 30]
})

# --- Empilhar verticalmente (axis=0) --- #
vendas_ano = pd.concat([vendas_jan, vendas_fev], axis=0, ignore_index=True)
vendas_ano
```

## O método `join()`: foco em índices

```python
# --- Colocar o ID do cliente como índice --- #
df_clientes_idx = df_clientes.set_index('ID_Cliente')
df_pedidos_idx = df_pedidos.set_index('Cod_Cliente')

# --- O .join() faz um left join por padrão --- #
df_resultado = df_clientes_idx.join(df_pedidos_idx)

df_resultado
```

|Comando|Quando usar|Chave de ligação|
|:-:|:-:|:-:|
|`pd.merge()`|Cruzamento complexo de tabelas relacionais|Colunas com nomes iguais ou diferentes|
|`pd.concat()`|Juntar arquivos com a mesma estrutura (ex: relatórios mensais)|Alinhamento por nome de coluna ou linha|
|`df.join()`|Quando você já transformou as chaves em índices das tabelas|Sempre através do Index|
