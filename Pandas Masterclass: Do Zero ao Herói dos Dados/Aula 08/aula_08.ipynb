## Preparando o Dataset

```python
# --- Importar o Pandas --- #
import pandas as pd

# --- Criar os dados --- #
dados = {
    'Loja': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Trimestre': ['Q1', 'Q2', 'Q1', 'Q2', 'Q1', 'Q2'],
    'Vendas': [1000, 1500, 800, 1200, 2000, 2100]
}

# --- Criar o DataFrame --- #
df = pd.DataFrame(dados)
df
```

## `pivot_table`: Tornando o Dado "Largo"

```python
# --- A pivot_table é como a tabela dinâmica no Excel --- #
df_pivot = df.pivot_table(
    values='Vendas',  # o que vai dentro da tabela
    index='Loja',  # quem será o index
    columns='Trimestre',  # quem será a coluna
    aggfunc='sum'  # a operação de agregação
)

# --- Mostrar a tabela pivotada --- #
df_pivot
```

##`melt`: Tornando o Dado "Longo"

```python
# --- "Derreter" a tabela --- #
df_melted = df_pivot.reset_index().melt(
    id_vars='Loja',  # index da tabela pivotada
    var_name='Trimestre',  # colunas a serem separadas
    value_name='Vendas'  # o que irá dentro da tabela
)

# --- Mostrar a tabela "derretida" --- #
df_melted
```

## `stack` e `unstack`: O Nível Hierárquico

- `stack`: Empilha as colunas para o índice (transforma **largo** em **longo**)
- `unstack`: Desempilha o índice para as colunas (transforma **longo** em **largo**)

```python
# --- Criar um índice hierárquico --- #
df_hierarquico = df.set_index(['Loja', 'Trimestre'])
df_hierarquico
```

```python
# --- Unstack: move "Trimestre" para as colunas --- #
df_unstacked = df_hierarquico.unstack()
df_unstacked
```

```python
# --- Stack: move "Trimestre" para o índice --- #
df_stacked = df_unstacked.stack()
df_stacked
```

|**Método**|**Objetico**|**Direção**|
|:-:|:-:|:-:|
|`pivot_table`|Criar relatórios comparativos|Longo -> Largo|
|`melt`|Preparar gráficos ou ML|Largo -> Longo|
|`unstack`|Mover índice para coluna|Longo -> Largo|
|`stack`|Mover coluna para índice|Largo -> Longo|
