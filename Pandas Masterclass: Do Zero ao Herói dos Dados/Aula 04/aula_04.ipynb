```python
# Importar o Pandas
import pandas as pd
```

## Preparando os dados

```python
# Criar os dados
dados = {
    'Nome': ['Alice', 'Bruno', 'Carla', 'Davi',
             'Eduarda', 'Fernando', 'Gabriela', 'Heitor'],
    'Departamento': ['TI', 'TI', 'TI', 'Vendas',
                     'Marketing', 'TI', 'Vendas', 'Marketing'],
    'Salário': [8500, 9200, 5000, 4200,
                6000, 10500, 3800, 7200],
    'Anos experiência': [5, 6, 3, 1,
                         4, 8, 2, 5],
    'Remoto': [True, False, False, True,
               True, False, False, True]
}

df = pd.DataFrame(dados).set_index('Nome')

df
```

## .iloc: Seleção por posição (coordenadas)

```python
# Selecionar a primeira linha e a primeira coluna
df.iloc[0, 0]  # linha, coluna
```

```python
# Selecionar as 3 primeiras linhas e todas as colunas
df.iloc[0:3, :]
```

```python
# Selecionar as colunas 1 e 2 para as linhas 4 a 6
df.iloc[4:7, 1:3]
```

## .loc: Seleção por rótulo (índice)

```python
# Selecionar todos os dados de Gabriela
df.loc['Gabriela']
```

```python
# Selecionar o salário e departamento
df.loc[['Bruno', 'Fernando'], ['Salário', 'Departamento']]
```

```python
# Fatiar o conjunto
df.loc['Alice': 'Eduarda', 'Salário']
```

## Máscara booleanas

```python
# Mostrar os dados a partir de uma condição
df_salario = df['Salário'] > 7000
df[df_salario]
```

```python
# Aplicar condições para a seleção dos dados (E)
filtro_e = (df['Departamento'] == 'TI') & (df['Remoto'] == True)
df[filtro_e]
```

```python
# Aplicar condições para a seleção dos dados (OU)
filtro_ou = (df['Salário'] < 5000) | (df['Anos experiência'] < 4)
df[filtro_ou]
```

## O método .query(): elegância SQL

```python
# Mais legível que por conector lógico
limite = 5000
busca = df.query('Salário > @limite and Departamento == "Marketing"')
busca
```
