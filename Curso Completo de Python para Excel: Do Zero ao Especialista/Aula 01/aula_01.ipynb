# Criar o Workbook

```python
# --- Importar a biblioteca --- #
import openpyxl

# --- Criar o workbook na memória --- #
wb = openpyxl.Workbook()

# --- Obter/ativar a planilha padrão do arquivo --- #
ws = wb.active

# --- Título da planilha --- #
ws.title = 'Relatório de Vendas'
```

# Adicionar informação

```python
# --- Escrita de informação em células de modo explícito --- #
ws['A1'] = 'ID Venda'
ws['B1'] = 'Valor Bruto'

# --- Escrita de informação em células por coordenadas --- #
ws.cell(row=2, column=1, value=1001)
ws.cell(row=2, column=2, value=450.75)
```

# Salvar o arquivo

```python
# --- Salvar da memória para um arquivo Excel --- #
wb.save('vendas_iniciais.xlsx')
print('Arquivo salvo com sucesso!')
```
